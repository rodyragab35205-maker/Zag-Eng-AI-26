#### 2) Create two arrays:

- one using a Python loop

- one using NumPy vectorization 

``to add 10 to each element.``

**Which one is faster and why?**

```python
#using Numpy
numpy_arr=np.arange(1,10)
print(numpy_arr)
print(numpy_arr+10)

#using python
python_arr=[1,2,3,4,5,6,7,8,9]
print(python_arr)
for n in python_arr:
    print(n+10)
```

```python
# Numpy is faster than python 
#because numpy use vectorize to perform operations (بتعملها مرة واحدة)
#but python makes loops to perform operation on element by element
```
