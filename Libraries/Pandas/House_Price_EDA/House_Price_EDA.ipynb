## Data Cleaning

```python
# Fill missing 'Price (in rupees)' with the mean
df['Price (in rupees)'].fillna(df['Price (in rupees)'].mean(),inplace=True)
```

```python
# Fill missing 'Carpet Area' with median (safer than mean for skewed data)
df['Carpet Area'].fillna(df['Carpet Area'].mode()[0],inplace=True)
```

```python
# Fill missing 'Status' with 'Unknown'
df['Status'].fillna('Unknown',inplace=True)
```

```python
# Drop columns with mostly missing values (>80%)
#>80% === >150024
#Dimensions,Plot Area
df.drop(columns=['Dimensions','Plot Area'],inplace=True)
```

## Aggregations & Relationships
